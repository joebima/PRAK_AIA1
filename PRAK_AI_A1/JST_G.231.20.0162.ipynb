```python
!pip install keras #>instalasi keras library
import keras #>memanggil library keras
import numpy as np #>memanggil library numpy
from sklearn.datasets import load_iris #>inisialisasi iris dari sklearn dataset dan memanggil iris
dataset=load_iris() #> fungsi load dataset iris dari proses sebelumnya
print (dataset) #> mencetak isi dataset
```

```python
print(type(dataset)) #> mencetak tipe dataset
print(len(dataset)) #> mencetak panjang dataset
print(dataset.keys()) 
print(dataset['data'][0:5])
print(dataset['target'][0:5])
print(dataset['target_names'][0:5])
print(dataset['DESCR'])
print(dataset['feature_names'])
print(dataset['filename'])
x=dataset['data']
print(len(x))
y=dataset['target']
print(len(y))
print(x[0])
print(y[0])
```

```python
from keras.utils import to_categorical
Ny=len(np.unique(y))
print(Ny)
Y=to_categorical(y,num_classes=Ny)
print(y[0:5])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,Y,test_size=0.10,shuffle=True)
print(x_train[0:5])
print(x_test[0:5])
print(y_train[0:5])
print(y_test[0:5])
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np
scaler=StandardScaler()
scaler.fit(x_train)

X_train=scaler.transform(x_train)
X_test=scaler.transform(x_test)
print(np.amax(X_train,axis=0))
print(np.amin(X_train,axis=0))
```

```python
!pip install tensorflow
import keras
from keras.models import Sequential
from keras.layers import Dense

model=Sequential()
model.add(Dense(20, input_dim=X_train.shape[1], activation='relu'))

keras.layers.Dropout(0.2)
model.add(Dense(20, activation='relu'))
keras.layers.Dropout(0.2)
model.add(Dense(20, activation='relu'))


model.add(Dense(3, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, epochs=195, batch_size=80,validation_split=0.1)
```

```python
print(model.evaluate(X_train, y_train))
```

```python
testing=model.evaluate(X_test, y_test)
#print("\n%s: %.2f%%" % (model.evaluate[1]+'uracy of Model on testing data', testing[1]*100))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
predictions = np.argmax(model.predict(X_test), axis=1)
Y_test = np.argmax(y_test, axis=1)
#to get confusion matrix
print(confusion_matrix(Y_test, predictions))
#to get values of all evalution metrics
print(classification_report(Y_test, predictions))
```
